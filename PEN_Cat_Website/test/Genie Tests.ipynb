# Create a page (Books Example)

```julia
# 1. Create Controller
Genie.REPL.new_controller("Books")

# 2. Test if the module is working by executing a functtion in it
using BooksController
BooksController.billgatesbooks() 
# file content: https://github.com/essenciary/Genie.jl (search for "BookController")

# 3. Create a link for the module/function
using Genie.Router
route("/bgbooks", BooksController.billgatesbooks)
# localhost:8000/bgbooks <- shows you the output of the fct billgatebooks
# But we want to have no HTML in the julia code, so we create a view file which has mostly HTML and can call julia

# 4. After creating the views folder in books/ and creating the html.jl file named after the function we call
# we get in touch
touch("app/resources/books/views/billgatesbooks.jl.html")
# file content: https://github.com/essenciary/Genie.jl (search for "<!-- billgatesbooks.jl.html -->")

# 5. We have to change the BooksController.jl file a bit, so that we can use the view:
#=
replace the billgatebooks function with:
function billgatesbooks()
  html!(:books, :billgatesbooks, books = BillGatesBooks)
end

and add
using Genie.Renderer

Now you are able to open the .jl.html file via localhost:8000/bgbooks
You can edit the view/html file as you like and add julia code in <% .. %> brackets

=#
```
